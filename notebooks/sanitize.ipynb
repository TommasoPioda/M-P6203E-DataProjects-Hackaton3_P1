```python
import json
from pathlib import Path

nb_path = Path("A_exploration_&_cleaning.ipynb")
nb = json.loads(nb_path.read_text(encoding="utf-8"))

# Cleans notebook metadata
nb["metadata"] = {}

# Cleans metadata of each cell
for cell in nb.get("cells", []):
    cell["metadata"] = {}

nb_path.with_suffix(".clean.ipynb").write_text(
    json.dumps(nb, indent=2, ensure_ascii=False),
    encoding="utf-8"
)

print("Cleaned notebook -> A_exploration_&_cleaning.clean.ipynb")
```
