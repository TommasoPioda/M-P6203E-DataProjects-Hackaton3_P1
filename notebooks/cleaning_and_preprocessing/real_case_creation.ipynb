```python
from utils.textual_utils.data_processing.data_utils import imbalance_classes, load_clean_citation_dataframe_from_files, normalize_list_text, normalize_text
import pandas as pd 
```

```python
test_pairs_df = pd.read_parquet("../../data/exploded_splits/test_pairs.parquet")
```

```python
imbalanced_df = imbalance_classes(
    test_pairs_df,
    target_col="is_reference_valid",
    strategy={
        0: 1_000_000,
        1: 19819,
    },
    allow_oversampling=True,
)
```

```python
imbalanced_df["keywords_article"] = imbalanced_df["keywords_article"].apply(normalize_list_text)
imbalanced_df["keywords_ref"] = imbalanced_df["keywords_ref"].apply(normalize_list_text)
imbalanced_df.fillna("", inplace=True)
imbalanced_df.to_parquet("../../data/exploded_splits/imbalanced_test_pairs.parquet", index=False)
```
