```python
import os
import pandas as pd
from utils import network_creation, target_definition, features_generation
import pyarrow.parquet as pq
import pyarrow as pa
```

```python
os.makedirs("data", exist_ok=True)

TRAIN_PATH = "data/split/train_pairs.parquet"
VAL_PATH = "data/split/validation_pairs.parquet"
TEST_PATH = "data/split/test_pairs.parquet"

# load the file 
train = pd.read_parquet(TRAIN_PATH)
val = pd.read_parquet(VAL_PATH)
test = pd.read_parquet(TEST_PATH)
```

# 1. Create the Network

```python
net_train = network_creation(train)
net_val = network_creation(val, "article_id", "ref_id", "is_reference_valid")
net_test = network_creation(test, "article_id", "ref_id", "is_reference_valid")
```

# 2. Generate the graph Features

```python
#final_train = features_generation(net_train, train_def)
final_val = features_generation(net_val, val, "article_id", "ref_id")
final_test = features_generation(net_test, test, "article_id", "ref_id")
```
