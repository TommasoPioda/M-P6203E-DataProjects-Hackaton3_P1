```python
import os
import pandas as pd
from utils import network_creation, target_definition, features_generation
```

```python
os.makedirs("data", exist_ok=True)

TRAIN_PATH = "data/split/train.parquet"
VAL_PATH = "data/split/validation.parquet"
TEST_PATH = "data/split/test.parquet"

# load the file 
train = pd.read_parquet(TRAIN_PATH)
val = pd.read_parquet(VAL_PATH)
test = pd.read_parquet(TEST_PATH)
```

```python
train = train.head(10000)
val = val.head(10000)
test = test.head(10000)
```

# 1. Create the Network

```python
net_train = network_creation(train)
net_val = network_creation(val)
net_test = network_creation(test)
```

# 2. Configure pairs of paper with target labels

```python
train_def = target_definition(net_train)
val_def = target_definition(net_val)
test_def = target_definition(net_test)
```

# 3. Generate the graph Features

```python
final_train = features_generation(net_train, train_def)
final_val = features_generation(net_val, val_def)
final_test = features_generation(net_test, test_def)
```

```python
print("final dataset for TRAIN set:")
display(final_train.head())
print("\nfinal dataset for VALIDATION set:")
display(final_val.head())
print("\nfinal dataset for TEST set:")
display(final_test.head())
```
